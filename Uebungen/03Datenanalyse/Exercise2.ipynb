### Import the necessary libraries

### Import the dataset & assign it to a variable called drinks

```python
!git clone https://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@github.com/HSD-W-MBA/BAB-WiSe-2022-2023.git
drinks = pd.read_csv('/content/BAB-WiSe-2022-2023/BAB_03_Datenanalyse_Python/exercises/exercise_2/drinks.csv')
drinks
```

### Which continent drinks more beer on average?

```python
drinks.groupby(by="continent")['beer_servings'].mean().sort_values(ascending = False)
```

### For each continent print the statistics for wine consumption.

```python
drinks.groupby(by="continent")['wine_servings'].describe()
```

### Print the mean alcohol consumption per continent for every column

```python
drinks.groupby(by="continent").mean()
```

### Print the median alcohol consumption per continent for every column

```python
drinks.groupby(by="continent").median()
```

### Print the mean, min and max values for spirit consumption per continent
#### This time output a DataFrame

```python
df = pd.DataFrame({"mean": drinks.groupby(by="continent")['spirit_servings'].mean(),
                   "min" : drinks.groupby(by="continent")['spirit_servings'].min(),
                   "max" : drinks.groupby(by="continent")['spirit_servings'].max()
})
df
```
